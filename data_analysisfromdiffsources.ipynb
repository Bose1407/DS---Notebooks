```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create synthetic data for Source 1 (CSV equivalent)
source_1 = pd.DataFrame({
    'id': range(1, 11),
    'sales': np.random.randint(500, 2000, 10),
    'profit': np.random.uniform(50, 300, 10),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries'], 10)
})

# Create synthetic data for Source 2 (Excel equivalent)
source_2 = pd.DataFrame({
    'id': range(1, 11),
    'rating': np.random.randint(1, 6, 10),
    'supplier': np.random.choice(['Supplier A', 'Supplier B', 'Supplier C'], 10)
})

# Create synthetic data for Source 3 (Database equivalent)
source_3 = pd.DataFrame({
    'id': range(1, 11),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 10),
    'customer_feedback': np.random.choice(['Positive', 'Neutral', 'Negative'], 10)
})

# Merge data from multiple sources
merged_data = pd.merge(source_1, source_2, on='id')
merged_data = pd.merge(merged_data, source_3, on='id')

# Display the merged data
print("Merged Data:")
print(merged_data)

# Fill missing values and remove duplicates
merged_data.fillna(0, inplace=True)
merged_data.drop_duplicates(inplace=True)

# Descriptive statistics
print("\nDescriptive Statistics:")
print(merged_data.describe())

# Grouped analysis
grouped_data = merged_data.groupby('category').agg({
    'sales': 'mean',
    'profit': 'mean',
    'rating': 'mean'
})
print("\nGrouped Data:")
print(grouped_data)

# Filtered data
filtered_data = merged_data[merged_data['sales'] > 1000]
print("\nFiltered Data where Sales > 1000:")
print(filtered_data)

# Visualization: Sales by Category
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='sales', data=merged_data)
plt.title('Sales by Category')
plt.show()

# Correlation heatmap for numeric columns
numeric_data = merged_data.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Boxplot for sales distribution by category
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='sales', data=merged_data)
plt.title('Sales Distribution by Category')
plt.show()



print("Data analysis and visualizations are complete.")
```
